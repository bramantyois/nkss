```python
import numpy as np

from nkss.examples.simpletube import SimpleTube
from nkss.utils import generate_ascending_sine, generate_sine_sweep
import matplotlib.pyplot as plt
```

```python
sr = 48000
n_sec = 0.2

freq = 100

max_amp = 16

triode = SimpleTube()
triode.calc_init_values(
    un=np.array([
        [-max_amp],
        [300]]),
    xnm1=np.array([
        [1.28e-7]]),
    vn=np.array([
        [-201.49],
        [1.45]]))

num_samples = int(sr * n_sec)
asc_sine = generate_ascending_sine(stop_amp=max_amp, sample_rate=sr)

out = triode.process_block(input_samples=asc_sine)

plt.subplot(2, 1, 1)
plt.plot(out)
plt.title('out')

plt.subplot(2, 1, 2)
plt.plot(asc_sine)
plt.title('sin')

plt.tight_layout()
plt.show()

```
