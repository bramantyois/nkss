```python
import numpy as np

from nkss.examples.hp_diode import HPDiode

import matplotlib.pyplot as plt
```

```python
sr = 44100
n_sec = 0.1

max_freq = 2000
min_freq = 10

diode = HPDiode()
diode.calc_init_values(
    un=np.array([[0]]),
    xnm1=1e-6*np.random.rand(2, 1),
    vn=1e-6*np.random.rand(1, 1))

num_samples = int(sr * n_sec)
freqs = np.linspace(2 * np.pi * min_freq, 2 * np.pi * max_freq, num_samples)
t = np.linspace(0, n_sec, num_samples)

sine_sweep = np.sin(freqs * t)

out = diode.process_block(input_samples=sine_sweep)

plt.subplot(2, 1, 1)
plt.plot(out)
plt.title('out')

plt.subplot(2, 1, 2)
plt.plot(sine_sweep)
plt.title('sin')

plt.tight_layout()
plt.show()
```
